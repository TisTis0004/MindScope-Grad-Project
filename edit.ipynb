```python
import json

input_path = "cache_windows/manifest.jsonl"
output_path = "cache_windows/manifest_fixed.jsonl"

with open(input_path, "r", encoding="utf-8") as f_in, open(output_path, "w", encoding="utf-8") as f_out:
    for line in f_in:
        line = line.strip()
        if not line:
            continue
        item = json.loads(line)
        item["pt_path"] = item["pt_path"].replace("..\\", "")
        f_out.write(json.dumps(item) + "\n")
```

```python
from data.dataloader import Loader

loader = Loader(
    ds="cache_windows/manifest.jsonl",
    batch_size=32,
    shuffle=False,
    num_workers=2,
    pin_memory=False,
    use_weighted_sampler=True,
).Loader()

sampled = []
for i, batch in enumerate(loader):
    sampled.extend(batch["y"].tolist())
    if i >= 100:   # about 101 batches, enough to inspect
        break

counter = Counter(sampled)
total = sum(counter.values())

print("Sampled distribution:")
for k in sorted(counter):
    print(f"class {k}: {counter[k]} ({counter[k] / total:.2%})")
```
