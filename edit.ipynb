```python
import json

input_path = "cache_windows/manifest.jsonl"
output_path = "cache_windows/manifest_fixed.jsonl"

with open(input_path, "r", encoding="utf-8") as f_in, open(output_path, "w", encoding="utf-8") as f_out:
    for line in f_in:
        line = line.strip()
        if not line:
            continue
        item = json.loads(line)
        item["pt_path"] = item["pt_path"].replace("..\\", "")
        f_out.write(json.dumps(item) + "\n")
```
